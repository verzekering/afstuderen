```python
import hydropandas as hpd
import os

organisation = "rotterdam"
auth = ("__key__", "D5aclEis.RBUeIvKA6jrLVUzNpPATjvGyLXsLAx7P")

extent = (46342, 144971, 409874, 462946) 

# Load data
oc = hpd.read_lizard(
    extent=extent,
    which_timeseries=["diver", "diver_validated", "hand"],
    organisation=organisation,
    auth=auth,
    tmin="2012-01-01",
)
```

```python
import pandas as pd
import os

# Bepaal Downloads-map (werkt op Windows)
downloads_dir = os.path.join(os.path.expanduser("~"), "Downloads")
out_file = os.path.join(downloads_dir, "groundwater_all_wells.csv")

rows = []

for well_code, row in oc.iterrows():
    gw = row["obs"]

    if gw is None or len(gw) == 0:
        continue

    df = gw.copy()
    df = df.reset_index()   # datum naar kolom
    df["well_code"] = well_code

    # metadata toevoegen
    df["x"] = row["x"]
    df["y"] = row["y"]
    df["screen_top"] = row["screen_top"]
    df["screen_bottom"] = row["screen_bottom"]
    df["ground_level"] = row["ground_level"]
    df["tube_nr"] = row["tube_nr"]

    rows.append(df)

# Alles samenvoegen
all_data = pd.concat(rows, ignore_index=True)

# Opslaan in Downloads
all_data.to_csv(out_file, index=False)

print(f"✅ Saved {len(all_data)} rows to:")
print(out_file)
```
